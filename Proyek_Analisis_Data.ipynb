## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Bagaimana pola penyewaan sepeda berdasarkan jam dalam sehari pada hari kerja dibandingkan hari libur selama tahun 2012, dan jam berapa yang memiliki rata-rata penyewaan tertinggi untuk masing-masing kategori hari?
- **Pertanyaan 2:** Bagaimana pengaruh suhu dan kelembaban terhadap jumlah penyewaan sepeda harian selama periode 2011-2012, dan pada kategori suhu mana (dingin, sedang, panas) penyewaan mencapai puncaknya?

### Assessing Data

#### Identifying ... problem

```python
# ============================================
# ASSESSING DATA
# ============================================

print("="*50)
print("ASSESSING DATA - day.csv")
print("="*50)

print("\nInformasi Dataset:")
day_df.info()

print("\nMissing Values:")
print(day_df.isnull().sum())

print(f"\nJumlah Duplikat: {day_df.duplicated().sum()}")

print("\nStatistik Deskriptif:")
display(day_df.describe())

print("\nPemeriksaan Outlier (Metode IQR):")
numeric_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
for col in numeric_cols:
    Q1 = day_df[col].quantile(0.25)
    Q3 = day_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = day_df[(day_df[col] < lower) | (day_df[col] > upper)].shape[0]
    print(f"{col:15s}: {outliers:3d} outliers (batas: {lower:.3f} - {upper:.3f})")

print("\nKonsistensi Nilai Kategorikal:")
print(f"season    : {sorted(day_df.season.unique())} -> valid")
print(f"yr        : {sorted(day_df.yr.unique())} -> valid")
print(f"mnth      : {sorted(day_df.mnth.unique())} -> valid")
print(f"holiday   : {sorted(day_df.holiday.unique())} -> valid")
print(f"weekday   : {sorted(day_df.weekday.unique())} -> valid")
print(f"workingday: {sorted(day_df.workingday.unique())} -> valid")
print(f"weathersit: {sorted(day_df.weathersit.unique())} -> valid")

print("\nIntegritas Data (casual + registered == cnt):")
check_day = (day_df['casual'] + day_df['registered'] == day_df['cnt']).all()
print(f"day.csv  : {check_day}")

# Assessing hour.csv
print("\n" + "="*50)
print("ASSESSING DATA - hour.csv")
print("="*50)

print("\nInformasi Dataset:")
hour_df.info()

print("\nMissing Values:")
print(hour_df.isnull().sum())

print(f"\nJumlah Duplikat: {hour_df.duplicated().sum()}")

print("\nPemeriksaan Outlier:")
for col in numeric_cols:
    if col in hour_df.columns:
        Q1 = hour_df[col].quantile(0.25)
        Q3 = hour_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = hour_df[(hour_df[col] < lower) | (hour_df[col] > upper)].shape[0]
        print(f"{col:15s}: {outliers:3d} outliers")

check_hour = (hour_df['casual'] + hour_df['registered'] == hour_df['cnt']).all()
print(f"\nIntegritas hour.csv: {check_hour}")
```

**Steps to Take:**
- Tidak ada missing values pada kedua dataset, sehingga tidak diperlukan tindakan imputasi.
- Tidak terdapat duplikasi data.
- Outlier yang terdeteksi merupakan variasi alami dari data penyewaan sepeda (misalnya, lonjakan pada jam sibuk dan penurunan pada dini hari), sehingga tidak perlu dihapus.
- Nilai kategorikal sudah konsisten dan sesuai dengan dokumentasi dataset.
- Data siap untuk masuk ke tahap cleaning untuk penyesuaian format dan penambahan kolom baru.

**Insight:**
- Dataset sudah cukup bersih dengan tidak adanya missing values maupun duplikasi.
- Outlier yang ditemukan bersifat wajar dan merepresentasikan variasi musiman dan pola jam sibuk.
- Integritas data terjaga dengan baik (casual + registered = cnt terpenuhi di kedua dataset).

### Cleaning Data

#### Fixing ... problem

```python
# ============================================
# CLEANING DATA
# ============================================

day_clean = day_df.copy()
hour_clean = hour_df.copy()

# 1. Konversi tipe data tanggal
day_clean['dteday'] = pd.to_datetime(day_clean['dteday'])
hour_clean['dteday'] = pd.to_datetime(hour_clean['dteday'])

# 2. Mapping nilai kategorikal untuk interpretasi yang lebih mudah
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
weather_map = {1: 'Cerah/Berawan', 2: 'Berkabut', 3: 'Hujan Ringan/Salju', 4: 'Hujan Lebat/Badai'}
day_map = {0: 'Senin', 1: 'Selasa', 2: 'Rabu', 3: 'Kamis', 4: 'Jumat', 5: 'Sabtu', 6: 'Minggu'}

day_clean['season_label'] = day_clean['season'].map(season_map)
hour_clean['season_label'] = hour_clean['season'].map(season_map)

day_clean['weather_label'] = day_clean['weathersit'].map(weather_map)
hour_clean['weather_label'] = hour_clean['weathersit'].map(weather_map)

day_clean['weekday_name'] = day_clean['weekday'].map(day_map)
hour_clean['weekday_name'] = hour_clean['weekday'].map(day_map)

# 3. Penanda tahun
day_clean['year_label'] = day_clean['yr'].map({0: '2011', 1: '2012'})
hour_clean['year_label'] = hour_clean['yr'].map({0: '2011', 1: '2012'})

# 4. Konversi nilai ternormalisasi ke nilai aktual
day_clean['temp_c'] = day_clean['temp'] * 41
hour_clean['temp_c'] = hour_clean['temp'] * 41
day_clean['hum_pct'] = day_clean['hum'] * 100
hour_clean['hum_pct'] = hour_clean['hum'] * 100
day_clean['windspeed_kph'] = day_clean['windspeed'] * 67
hour_clean['windspeed_kph'] = hour_clean['windspeed'] * 67

# 5. Kategorisasi suhu untuk analisis binning
day_clean['temp_category'] = pd.cut(
    day_clean['temp_c'],
    bins=[0, 10, 20, 30, 50],
    labels=['Dingin (<10 C)', 'Sejuk (10-20 C)', 'Hangat (20-30 C)', 'Panas (>30 C)']
)

day_clean['hum_category'] = pd.cut(
    day_clean['hum_pct'],
    bins=[0, 40, 70, 101],
    labels=['Rendah (<40%)', 'Sedang (40-70%)', 'Tinggi (>70%)']
)

print("Cleaning data selesai.")
print(f"day_clean  : {day_clean.shape}")
print(f"hour_clean : {hour_clean.shape}")
print("\nKolom baru yang ditambahkan:")
print("- season_label, weather_label, weekday_name")
print("- year_label")
print("- temp_c, hum_pct, windspeed_kph")
print("- temp_category, hum_category")
```

**Insight:**
- Dilakukan konversi nilai ternormalisasi ke nilai aktual (suhu dalam Celcius, kelembaban dalam persen) untuk interpretasi yang lebih intuitif.
- Nilai kategorikal telah dimapping ke label yang deskriptif.
- Binning awal untuk suhu dan kelembaban telah siap untuk digunakan dalam tahap EDA.

## Exploratory Data Analysis (EDA)

```python
# ============================================
# EXPLORATORY DATA ANALYSIS
# ============================================

print("="*60)
print("STATISTIK DESKRIPTIF - DATA HARIAN")
print("="*60)
display(day_clean.describe())

print("\nRATA-RATA PENYEWAAN PER TAHUN:")
yearly_stats = day_clean.groupby('year_label')['cnt'].agg(['mean', 'sum', 'max'])
display(yearly_stats)

print("\nRATA-RATA PENYEWAAN PER MUSIM:")
season_stats = hour_clean.groupby('season_label')['cnt'].describe()
display(season_stats)

# --- EDA Pertanyaan 1: Pola Jam ---
print("\n" + "="*60)
print("EDA PERTANYAAN 1: POLA PENYEWAAN PER JAM (2012)")
print("="*60)

hour_2012 = hour_clean[hour_clean['yr'] == 1].copy()
hour_2012['day_type'] = hour_2012['workingday'].map({1: 'Hari Kerja', 0: 'Hari Libur'})

hourly_pattern = hour_2012.groupby(['hr', 'day_type'])['cnt'].mean().reset_index()
pivot_pattern = hourly_pattern.pivot(index='hr', columns='day_type', values='cnt')

display(pivot_pattern)

print("\nJam dengan rata-rata penyewaan tertinggi:")
for dtype in ['Hari Kerja', 'Hari Libur']:
    top = hourly_pattern[hourly_pattern['day_type'] == dtype].nlargest(3, 'cnt')
    print(f"{dtype}:")
    for _, row in top.iterrows():
        print(f"  Jam {int(row['hr']):02d}:00 -> {row['cnt']:.0f} penyewaan")

# --- EDA Pertanyaan 2: Suhu & Kelembaban ---
print("\n" + "="*60)
print("EDA PERTANYAAN 2: PENGARUH SUHU DAN KELEMBABAN")
print("="*60)

print("\nKorelasi dengan jumlah penyewaan harian (cnt):")
for col in ['temp', 'hum', 'windspeed']:
    corr = day_clean[col].corr(day_clean['cnt'])
    print(f"  {col:15s}: {corr:.4f}")

print("\nRata-rata penyewaan per kategori suhu:")
temp_stats = day_clean.groupby('temp_category', observed=True)['cnt'].describe()
display(temp_stats)

print("\nRata-rata penyewaan per kategori kelembaban:")
hum_stats = day_clean.groupby('hum_category', observed=True)['cnt'].describe()
display(hum_stats)
```

**Insight:**
- Pertanyaan 1: Pada hari kerja, pola penyewaan menunjukkan dua puncak (bimodal) yaitu jam 08:00 (pagi) dan jam 17:00-18:00 (sore), yang mengindikasikan pola komuter. Pada hari libur, pola cenderung unimodal dengan puncak di siang hari (jam 12:00-15:00), mengindikasikan pola rekreasi.
- Pertanyaan 2: Suhu memiliki korelasi positif sebesar +0.40 terhadap jumlah penyewaan, sedangkan kelembaban berkorelasi negatif sebesar -0.32. Penyewaan tertinggi terjadi pada suhu hangat (20-30 C) dengan rata-rata harian ~5.000 penyewaan.

## Analisis Lanjutan (Opsional)

```python
# ============================================
# ANALISIS LANJUTAN: CLUSTERING & BINNING
# ============================================

# --- 1. Time-based Clustering ---
print("="*60)
print("TIME-BASED CLUSTERING")
print("="*60)

def get_time_cluster(hr):
    if hr in [0, 1, 2, 3, 4, 5]:
        return 'Dini Hari (0-5)'
    elif hr in [6, 7, 8, 9]:
        return 'Pagi/Morning Rush (6-9)'
    elif hr in [10, 11, 12, 13, 14, 15]:
        return 'Siang (10-15)'
    elif hr in [16, 17, 18, 19]:
        return 'Sore/Evening Rush (16-19)'
    else:
        return 'Malam (20-23)'

hour_clean['time_cluster'] = hour_clean['hr'].apply(get_time_cluster)

cluster_avg = hour_clean.groupby('time_cluster', observed=True)['cnt'].agg(['mean', 'median', 'count'])
display(cluster_avg)

# Komposisi casual vs registered per cluster
print("\nKomposisi pengguna per cluster waktu:")
cluster_user = hour_clean.groupby('time_cluster', observed=True)[['casual', 'registered']].mean()
cluster_user['casual_pct'] = (cluster_user['casual'] / (cluster_user['casual'] + cluster_user['registered']) * 100).round(1)
cluster_user['registered_pct'] = (cluster_user['registered'] / (cluster_user['casual'] + cluster_user['registered']) * 100).round(1)
display(cluster_user)

# --- 2. Binning Analysis ---
print("\n" + "="*60)
print("BINNING ANALYSIS")
print("="*60)

# Temperature binning detail
temp_bins = pd.cut(
    hour_clean['temp_c'],
    bins=[0, 5, 10, 15, 20, 25, 30, 35, 45],
    labels=['<5C', '5-10C', '10-15C', '15-20C', '20-25C', '25-30C', '30-35C', '>35C']
)
temp_stats = hour_clean.groupby(temp_bins, observed=True)['cnt'].agg(['mean', 'median', 'count'])
print("\nRata-rata penyewaan per rentang suhu:")
display(temp_stats)

# Humidity binning detail
hum_bins = pd.cut(
    hour_clean['hum_pct'],
    bins=[0, 25, 50, 75, 101],
    labels=['Sangat Rendah (<25%)', 'Rendah (25-50%)', 'Sedang (50-75%)', 'Tinggi (>75%)']
)
hum_stats = hour_clean.groupby(hum_bins, observed=True)['cnt'].agg(['mean', 'median', 'count'])
print("\nRata-rata penyewaan per rentang kelembaban:")
display(hum_stats)

# --- 3. Usage Level Segmentation ---
print("\n" + "="*60)
print("SEGMENTASI TINGKAT PENGGUNAAN")
print("="*60)

quartiles = hour_clean['cnt'].quantile([0.25, 0.5, 0.75]).values
print(f"Q1 (25%): {quartiles[0]:.0f}")
print(f"Median  : {quartiles[1]:.0f}")
print(f"Q3 (75%): {quartiles[2]:.0f}")

def get_usage_level(cnt):
    if cnt <= quartiles[0]:
        return 'Rendah'
    elif cnt <= quartiles[2]:
        return 'Sedang'
    else:
        return 'Tinggi'

hour_clean['usage_level'] = hour_clean['cnt'].apply(get_usage_level)
print("\nDistribusi level penggunaan:")
print(hour_clean['usage_level'].value_counts())

print("\nTabulasi silang: Cluster Waktu vs Level Penggunaan:")
cross_tab = pd.crosstab(hour_clean['time_cluster'], hour_clean['usage_level'], normalize='index') * 100
display(cross_tab.round(1))

# --- 4. Casual vs Registered Analysis ---
print("\n" + "="*60)
print("ANALISIS CASUAL VS REGISTERED")
print("="*60)

total_casual = hour_clean['casual'].sum()
total_registered = hour_clean['registered'].sum()
total_all = total_casual + total_registered
print(f"Proporsi Casual    : {total_casual/total_all*100:.1f}%")
print(f"Proporsi Registered: {total_registered/total_all*100:.1f}%")

# Visualisasi: Bar chart per cluster waktu
cluster_viz = cluster_avg.reset_index()
cluster_viz = cluster_viz.sort_values('mean', ascending=True)

fig1 = go.Figure(go.Bar(
    x=cluster_viz['mean'],
    y=cluster_viz['time_cluster'],
    orientation='h',
    marker_color=['#6A994E', '#2E86AB', '#A23B72', '#F18F01', '#C73E1D'][::-1],
    text=cluster_viz['mean'].round(0),
    textposition='outside'
))

fig1.update_layout(
    title='Rata-rata Penyewaan per Cluster Waktu',
    xaxis_title='Rata-rata Penyewaan',
    yaxis_title='',
    template='plotly_white',
    width=800,
    height=400
)

fig1.show()

# Visualisasi: Casual vs Registered per jam
hourly_user = hour_clean.groupby('hr')[['casual', 'registered']].mean().reset_index()

fig2 = go.Figure()
fig2.add_trace(go.Scatter(
    x=hourly_user['hr'], y=hourly_user['registered'],
    mode='lines+markers', name='Registered',
    line=dict(width=3, color='#3498DB'), marker=dict(size=8)
))
fig2.add_trace(go.Scatter(
    x=hourly_user['hr'], y=hourly_user['casual'],
    mode='lines+markers', name='Casual',
    line=dict(width=3, color='#E67E22'), marker=dict(size=8)
))
fig2.update_layout(
    title='Rata-rata Pengguna Casual vs Registered per Jam',
    xaxis_title='Jam', yaxis_title='Rata-rata Jumlah Pengguna',
    xaxis=dict(tickmode='linear', tick0=0, dtick=2),
    template='plotly_white', width=900, height=450
)
fig2.show()
```

**Insight:**
- Time-based Clustering: Cluster Sore/Evening Rush (16-19) memiliki rata-rata penyewaan tertinggi (378 per jam), diikuti oleh Siang (230) dan Pagi/Morning Rush (217). Dini Hari (0-5) memiliki rata-rata terendah (25 per jam).
- Binning Analysis: Suhu optimal untuk penyewaan adalah pada rentang 25-35 C. Kelembaban optimal pada tingkat rendah hingga sedang (<70%).
- Segmentasi Pengguna: Registered users mendominasi di jam sibuk pagi dan sore (pola komuter), sedangkan casual users memiliki proporsi lebih tinggi di siang hari dan akhir pekan (pola rekreasi).
